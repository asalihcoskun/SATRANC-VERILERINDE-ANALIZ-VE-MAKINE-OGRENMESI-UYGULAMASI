```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
carlsen = pd.read_parquet("Carlsen V2.parquet")
nakamura = pd.read_parquet("Nakamura V2.parquet")
caruana = pd.read_parquet("Caruana V2.parquet")
```

```python
carlsen.head()
```

```python
df = pd.concat([carlsen, nakamura, caruana])
df.head()
```

```python
top_6_eco = df['ECO'].value_counts().head(6)

plt.figure(figsize=(13, 6))

plt.subplot(1, 2, 1)
top_6_eco.plot(kind='bar', color='skyblue')
plt.title('En ok Oynanan 6 ECO Kodu')
plt.xlabel('ECO Kodu')
plt.ylabel('Oyun Says')
plt.xticks(rotation=45)
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.pie(top_6_eco, labels=top_6_eco.index, autopct='%1.1f%%', startangle=120)
plt.title('En ok Oynanan 6 ECO Kodunun Oransal Dalm')
plt.axis('equal')
plt.tight_layout()

plt.show()
```

```python
GMs = ['Nakamura,Hi', 'Carlsen,M', 'Caruana,F']
```

```python
for player in GMs:
    player_games = df[(df['White'] == player) | (df['Black'] == player)]
    openings_count = player_games['ECO'].value_counts()
    most_played_opening = openings_count.idxmax()  # En ok oynanan al
    print(f"{player}'n en ok oynad al: {most_played_opening} - {openings_count.max()} oyun")
```

```python
def grafik_1 (player_name):
    player_games = df[(df['White'] == player_name) | (df['Black'] == player_name)]
    openings_count = player_games['ECO'].value_counts().head(5)
    game_class_count = player_games['Game Class'].value_counts()
    
    plt.figure(figsize=(15,10))

    plt.subplot(2, 2, 1)
    plt.barh(openings_count.index, openings_count)
    plt.title(f'{player_name} - En ok Oynad 5 Al')
    plt.xlabel('Oyun Says')
    plt.ylabel('Al')
    plt.tight_layout()

    plt.subplot(2, 2, 2)
    plt.pie(openings_count, labels=openings_count.index, autopct='%1.1f%%', startangle=120)
    plt.title(f'{player_name} - En ok Oynad 5 Aln Oransal Dalm')
    plt.axis('equal')
    plt.tight_layout()
    
    plt.subplot(2, 2, 3)
    plt.barh(game_class_count.index, game_class_count)
    plt.title(f'{player_name} - Oyun Snflar')
    plt.xlabel('Oyun Says')
    plt.ylabel('Oyun Snf')  # ylabel deitirildi
    plt.tight_layout()

    plt.subplot(2, 2, 4)
    plt.pie(game_class_count, labels=game_class_count.index, autopct='%1.1f%%', startangle=120)
    plt.title(f'{player_name} - Oyun Snflarnn Oransal Dalm')
    plt.axis('equal')
    plt.tight_layout()
    
    plt.show()
```

```python
grafik_1('Carlsen,M')
```

```python
grafik_1('Nakamura,Hi')
```

```python
grafik_1('Caruana,F')
```

```python
def grafik_2 (player_name):
    player_games = df[(df['White'] == player_name) | (df['Black'] == player_name)]
    oyun_uzunlugu_count=player_games['Game Length'].value_counts()
   
    plt.figure(figsize=(12,4))

    plt.subplot(1, 2, 1)
    plt.barh(oyun_uzunlugu_count.index, oyun_uzunlugu_count)
    plt.title(f'{player_name} - Oyunlarn Uzunluklarna Gre Saysal Dalm')
    plt.xlabel('Oyun Says')
    plt.ylabel('Oyun Uzunluu')  # ylabel deitirildi
    plt.tight_layout()

    plt.subplot(1, 2, 2)
    plt.pie(oyun_uzunlugu_count, labels=oyun_uzunlugu_count.index, autopct='%1.1f%%', startangle=120)
    plt.title(f'{player_name} - Oyunlarn Uzunluklarna Gre Oransal Dalm')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
    
def grafik_3 (player_name):  
    player_games = df[(df['White'] == player_name) | (df['Black'] == player_name)]
    best_move_rate_count=player_games['Best Move Rate Classify'].value_counts()
    
    plt.figure(figsize=(12,4))
    
    plt.subplot(1, 2, 1)
    plt.barh(best_move_rate_count.index, best_move_rate_count)
    plt.title(f'{player_name} - En iyi Hamleyi Oynama Oranlarnn Saysal Dalm')
    plt.xlabel('Oyun Says')
    plt.ylabel('En yi Hamle Oynama Oran')  # ylabel deitirildi
    plt.tight_layout()

    plt.subplot(1, 2, 2)
    plt.pie(best_move_rate_count, labels=best_move_rate_count.index, autopct='%1.1f%%', startangle=120)
    plt.title(f'{player_name} - En iyi Hamleyi Oynama Oranlarnn Oransal Dalm')
    plt.axis('equal')
    plt.tight_layout()
    
    plt.show()
```

```python
grafik_2('Carlsen,M')
```

```python
grafik_2('Nakamura,Hi')
```

```python
grafik_2('Caruana,F')
```

```python
grafik_3('Nakamura,Hi')
```
